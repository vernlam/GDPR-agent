```python
import mlflow

# Get your experiment
experiment = mlflow.get_experiment_by_name("/Shared/gdpr-agent-ci-evaluation")
print(f"Experiment ID: {experiment.experiment_id}")

# Get latest run
runs = mlflow.search_runs(
    experiment_ids=[experiment.experiment_id],
    order_by=["start_time DESC"],
    max_results=1
)

if not runs.empty:
    run_id = runs.iloc[0]['run_id']
    print(f"Latest run ID: {run_id}")
    
    # List artifacts
    client = mlflow.tracking.MlflowClient()
    artifacts = client.list_artifacts(run_id)
    
    print("\nArtifacts in this run:")
    for artifact in artifacts:
        print(f"  - {artifact.path}")
else:
    print("No runs found")
```

```python
import mlflow
import json

# Get the specific trace
client = mlflow.tracking.MlflowClient()

trace_id = "tr-358bfbe2d0df0283f0ec40e96fa0d6a4"

try:
    # Get trace by ID
    trace = client.get_trace(trace_id)
    
    print(f"Trace ID: {trace.info.request_id}")
    print(f"Status: {trace.info.status}")
    print(f"\n{'='*80}\n")
    
    # Iterate through all spans
    for span in trace.data.spans:
        if "retrieve" in span.name.lower() or "search" in span.name.lower():
            print(f"🔍 Span: {span.name}")
            print(f"   Type: {span.span_type}")
            print(f"   Status: {span.status}")
            
            # Print full inputs
            if span.inputs:
                print(f"\n   📥 INPUTS:")
                print(json.dumps(span.inputs, indent=2))
            
            # Print full outputs (NOT TRUNCATED)
            if span.outputs:
                print(f"\n   📤 FULL OUTPUTS:")
                print(json.dumps(span.outputs, indent=2))
            
            # Print attributes
            if span.attributes:
                print(f"\n   🏷️  ATTRIBUTES:")
                print(json.dumps(span.attributes, indent=2))
            
            print(f"\n{'-'*80}\n")
            
except Exception as e:
    print(f"❌ Error fetching trace: {e}")
    print(f"\nTrying alternative method...")
```

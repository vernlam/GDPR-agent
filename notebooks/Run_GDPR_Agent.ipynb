```python
# Debug: Show retrieved chunks from Cell 5
print("="*80)
print("📚 RETRIEVED CHUNKS BREAKDOWN")
print("="*80)

context = final_state["retrieved_context"]

# Split by the separator used in nodes.py
chunks = context.split("\n\n---\n\n")

for i, chunk in enumerate(chunks, 1):
    print(f"\n🔍 Chunk {i}/{len(chunks)}")
    print("-" * 80)
    # Extract source
    if "[SOURCE:" in chunk:
        source_end = chunk.find("]")
        source = chunk[1:source_end]
        content = chunk[source_end+1:].strip()
        print(f"📂 {source}")
        print(f"📄 Content (first 500 chars):")
        print(content[:500])
        if len(content) > 500:
            print(f"\n... ({len(content)} total characters)")
    else:
        print(chunk[:500])
    print()

print("="*80)
print(f"📊 Total chunks retrieved: {len(chunks)}")
print(f"📊 Total context length: {len(context)} characters")
print("="*80)
```
